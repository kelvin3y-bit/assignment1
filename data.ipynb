```python

import pandas as pd

# Load data
df = pd.read_csv("owid-covid-data.csv")

# Preview
print(df.columns)
print(df.head())
print(df.info())
print(df.isnull().sum())

```

```python
# Filter key countries
countries = ['Kenya', 'United States', 'India']
df = df[df['location'].isin(countries)]

# Convert date to datetime
df['date'] = pd.to_datetime(df['date'])

# Fill or interpolate missing values
df = df.fillna(0)  # or use df.interpolate()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot total cases
plt.figure(figsize=(12,6))
for country in countries:
    subset = df[df['location'] == country]
    plt.plot(subset['date'], subset['total_cases'], label=country)
plt.title("Total COVID-19 Cases Over Time")
plt.legend()
plt.show()
```

```python
df['death_rate'] = df['total_deaths'] / df['total_cases']
```

```python
sns.heatmap(df[['total_cases', 'total_deaths', 'total_vaccinations']].corr(), annot=True)
```

```python
# Plot vaccinations
plt.figure(figsize=(12,6))
for country in countries:
    subset = df[df['location'] == country]
    plt.plot(subset['date'], subset['people_fully_vaccinated'], label=country)
plt.title("Vaccination Progress")
plt.legend()
plt.show()
```

```python
import plotly.express as px

# Latest data only
latest = df[df['date'] == df['date'].max()]
fig = px.choropleth(latest,
                    locations="iso_code",
                    color="total_cases",
                    hover_name="location",
                    color_continuous_scale="Reds",
                    title="Global COVID-19 Case Distribution")
fig.show()
```

### Insights:
1. The US had the highest total cases by December 2022.
2. India showed a steady vaccination rollout starting in early 2021.
3. Kenya experienced a slower but steady increase in total vaccinations.
4. The death rate in Country X spiked during [Month/Year].
